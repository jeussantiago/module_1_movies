change "gross" and "budget" type to int to be aggregable

```python
def turn_money(x):
    try:
        x = int(x.replace(",", "").strip("$"))
    except:
        x = None
    return x

df2['gross'] = df2['gross'].map(lambda dollars: turn_money(dollars))
df2['budget'] = df2['budget'].map(lambda dollars: turn_money(dollars))

df2
```
