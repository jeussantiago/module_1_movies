```python
#The following code is used to import the csv file created from the API data and clean it
#using "Exploratory Data Analysis" methods
import pandas as pd
import ast
import json
from datetime import datetime
```

```python
#reads the data
df = pd.read_csv('list_of_popular_movies_1_to_19793.csv', engine='python')
df.tail()
```

```python
df.columns.values
```

```python
#creates a database using only the relevant columns we need
movies_df = df[['id', 'title', 'release_date', 'runtime', 'revenue', 'budget', 'genres', 'status','original_language','production_countries','production_companies', 'popularity','vote_average', 'vote_count','imdb_id']]
movies_df.head()
```

```python
print(movies_df.shape)
```

```python
#Gets rid of duplicate rows
movies_df = movies_df.drop_duplicates(keep='first')
```

```python
print(movies_df.shape)
```

```python
#creates a new column called release year that convets the date string to a numerical datatype that can be used for filtering

movies_df.release_date = pd.to_datetime(movies_df['release_date'])
movies_df['release_year'] = movies_df['release_date'].dt.year
movies_df.head()
```

```python
#A function that extracts values from strings given a key

def turn_string_into_list(pd_string, key):
    try:
        list_of_items = []
        string_list_of_dicts = ast.literal_eval(pd_string)
        for dict_pair in string_list_of_dicts:
            list_of_items.append(dict_pair[key])
            
        return list_of_items
        
    except:
        return None
```

```python
#A test of the above function to see if it could extract genres
turn_string_into_list(movies_df['genres'][1], 'name')
```

```python
#Creates new columns based on the extracted values
movies_df['genres_list'] = movies_df['genres'].map(lambda x: turn_string_into_list(x, 'name'))
movies_df['production_company_list'] = movies_df['production_companies'].map(lambda x: turn_string_into_list(x, 'name'))
movies_df['production_country_list'] = movies_df['production_countries'].map(lambda x: turn_string_into_list(x, 'name'))

movies_df.head()
```

```python
#Drops the old genres column
movies_df = movies_df.drop('genres', axis=1)
movies_df = movies_df.drop('production_companies', axis=1)
movies_df = movies_df.drop('production_countries', axis=1)
movies_df.head()
```

```python
movies_df.dtypes
```

```python
#Converts the popularity ranking and budget columns to float64 datatypes

movies_df['budget'] = movies_df['budget'].convert_objects(convert_numeric=True)
movies_df['popularity'] = movies_df['popularity'].convert_objects(convert_numeric=True)

movies_df.dtypes
```

```python
movies_df.head()
```

```python
#Checks to see how many rows contain Nan values
movies_df.isna().sum()
```

```python
#Drops all Nan values in the genres_list column

movies_df = movies_df.dropna(subset=['genres_list'], how='all')
```

```python
movies_df
```

```python
all_genres = ['Action', 'Adventure', 'Animation', 'Comedy', 'Crime', 'Documentary',
              'Drama', 'Family', 'Fantasy', 'History', 'Horror', 'Music', 'Mystery',
              'Romance', 'Science Fiction', 'TV Movie', 'Thriller', 'War', 'Western']
```

```python
#Creates a new column per genre and sets the row values to zero or one, we use this for boolean filtering later

genre_matrix = pd.get_dummies(movies_df['genres_list'].apply(pd.Series).stack()).sum(level=0)
```

```python
#get a list of cols not officially a genre "all_genre"
drop_genre_cols= [genre for genre in list(genre_matrix.columns) if genre not in all_genres]
drop_genre_cols
```

```python
#drop the columns we don't need
genre_matrix.drop(drop_genre_cols, axis=1, inplace=True)

```

```python
genre_matrix
```

```python
movies_df
```

```python
#merge the two dataframes by the indexes
movies_df = movies_df.join(genre_matrix)
```

```python
#fill null values in genres as 0
for genre in all_genres:
    movies_df[genre].fillna(0, inplace=True)
```
