```python
import requests
import json
import pandas as pd
import time

from config import api_key

url = 'https://api.themoviedb.org/3/movie/550?api_key={}'.format(api_key)

response = requests.get(url)
print(response)
```

```python
#Proves that the list of trending movies from the api contains unique
#values by creating a list of id dictionaries and comparing it to a
#set of that list

#A list holder for the id-value pairs
id_list = []

#A count of the number of attempts to receive id information from the API
id_count = 0

#The URL page number for the movie information from the API
id_page = 1

#There are 991 pages of movies on TMDB's trending list
while id_page <= 991:
    
    id_count+=1
    url_popular = 'https://api.themoviedb.org/3/movie/popular?api_key={}&language=en-US&page={}'.format(api_key, id_page)
    response = requests.get(url_popular)
    
    if response.status_code == 200 and type(response.json()['results']) == list:
        twenty_movies = response.json()['results']
        
        for movie in twenty_movies:
            id_list.append(movie['id'])
            
        id_page+=1
        if id_page %5==0:
            print('{} movie ids have been added so far.'.format(id_page *20))
    elif response.status_code == 404:
        break
    else:
        print(response.status_code)
    time.sleep(.5)
```

```python
#If the legnth of the movie id list and the legnth of the set
#of the movie list are equal then the list contains unique values

print(len(id_list))
print(len(set(id_list)))
```

```python
#Returns the larger movie dictionary when a smaller dictionary
#is given as an argument

def find_movie_info(movie):
    movie_id = movie['id']
    url_movie ='https://api.themoviedb.org/3/movie/{}?api_key={}&language=en-US'.format(movie_id, api_key)
    response = requests.get(url_movie)
    time.sleep(.25)
    return response.json()
```

```python
#A blank holder for the large list of dictionaries
most_popular_movies_dict_list = []
```

```python
#The following code creates a list of expanded movie dictionaries 

dict_count = 0
dict_page = 301
while len(most_popular_movies_dict_list) < len(set(id_list)) and dict_count < 2*round(len(set(id_list))/20) and dict_page < 991:
    
    dict_count+=1
    url_popular = 'https://api.themoviedb.org/3/movie/popular?api_key={}&language=en-US&page={}'.format(api_key, dict_page)
    response = requests.get(url_popular)
    
    if response.status_code == 200:
        twenty_movies = response.json()['results']
        
        for movie in twenty_movies:
            full_dict = find_movie_info(movie)
            most_popular_movies_dict_list.append(full_dict)
        
        if dict_page %5==0:
            print('{} movies have been added so far. There have been {} attempts'.format(len(most_popular_movies_dict_list), dict_count))
        dict_page+=1
    elif response.status_code == 404:
        break
    else:
        print(response.status_code)
    time.sleep(.25)
    
```

```python
df = pd.DataFrame(most_popular_movies_dict_list)
print(len(df))
df.tail()
```

```python
df.to_csv("list_of_popular_movies_1_to_19793.csv", index=False)
```
