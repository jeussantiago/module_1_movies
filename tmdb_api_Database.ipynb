```python
def find_movie_info(movie):
    movie_id = movie['id']
    url_movie ='https://api.themoviedb.org/3/movie/{}?api_key={}&language=en-US'.format(movie_id, api_key)
    response = requests.get(url_movie)
    time.sleep(.25)
    return response.json()
```

```python
id_list = []
count = 0
page = 1
while page <= 991:
    count+=1
    url_popular = 'https://api.themoviedb.org/3/movie/popular?api_key={}&language=en-US&page={}'.format(api_key, page)
    response = requests.get(url_popular)
    
    if response.status_code == 200:
        twenty_movies = response.json()['results']
        
        for movie in twenty_movies:
            id_list.append(movie['id'])
        page+=1
        if page %5==0:
            print('{} movie ids have been added so far.'.format(page *20))
    elif response.status_code == 404:
        break
    else:
        print(response.status_code)
    time.sleep(.5)
```
